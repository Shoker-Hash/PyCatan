```python
def calcula_BeneficiosCoste(arrayCostes, arrayBeneficios):
    i = 0
    resultado = 0
    for coste in arrayCostes:
        if coste<0:
            resultado = resultado + arrayBeneficios[i]/(-coste)
        i = i+1
    return resultado
```

```python
def compra_objetivo(self, turnoActual, board):
    #Variables Alg Genetico = GustoCiudad,GustoPoblado,GustoCarretera,GustoDesarrollo,timechange,timechange2,timechange3, importanciaTiempo
    townPossibilities = self.board.valid_town_nodes(self.id)
    cityPossibilities = self.board.valid_city_nodes(self.id)
    materialesCiudad = self.hand.resources.has_this_more_materials(BuildConstants.CITY)
    Vt = self.__CR__(board.nodes)[self.id]
    if cityPossibilities and townPossibilities:
        Cc, arrayCosteCiudad = materialesNecesarios(BuildConstants.CITY, self.hand.resources)
        Cp, arrayCostePueblo = materialesNecesarios(BuildConstants.TOWN, self.hand.resources)
        Cc = calcula_BeneficiosCoste(arrayCosteCiudad, Vt)
        Cp = calcula_BeneficiosCoste(arrayCostePueblo, Vt)
        result = GustoCiudad * Cc   - GustoPoblado * Cp  -[ 1 - (turnoActual*1/35*time_change) ]*importancia_Tiempo
        if result>0:
            return BuildConstants.CITY
        else: 
            return BuildConstants.TOWN
    elif townPossibilities:
        Cd, arrayCosteDesarrollo = materialesNecesarios(BuildConstants.CARD, self.hand.resources)
        Cp, arrayCostePueblo = materialesNecesarios(BuildConstants.TOWN, self.hand.resources)
        Cd = calcula_BeneficiosCoste(arrayCosteDesarrollo, Vt)
        Cp = calcula_BeneficiosCoste(arrayCostePueblo, Vt)
        result = GustoCiudad * Cd   - GustoPoblado * Cp  -[ 1 - (turnoActual*1/35*time_change2) ]*importancia_Tiempo
        if result>0:
            return BuildConstants.CARD
        else: 
            return BuildConstants.TOWN
    elif cityPossibilities:
        carreteraConstruible = 0
        if len(board.valid_road_nodes()) > 0:
            carreteraConstruible = 1
        Cc, arrayCosteCiudad = materialesNecesarios(BuildConstants.CITY, self.hand.resources)
        Cr, arrayCosteCarretera = materialesNecesarios(BuildConstants.ROAD, self.hand.resources)
        Cc = calcula_BeneficiosCoste(arrayCosteCiudad, Vt)
        Cr = calcula_BeneficiosCoste(arrayCosteCarretera, Vt)
        result = GustoCiudad * Cc   - GustoPoblado*Cr*carreteraConstruible -[ 1 - (turnoActual*1/35*time_change3) ]*importancia_Tiempo
        if result>0:
            return BuildConstants.CITY
        else:
            return BuildConstants.ROAD
    else:
        carreteraConstruible = 0
        if len(board.valid_road_nodes()) > 0:
            carreteraConstruible = 1
        Cd, arrayCosteDesarrollo = materialesNecesarios(BuildConstants.CARD, self.hand.resources)
        Cr, arrayCosteCarretera = materialesNecesarios(BuildConstants.ROAD, self.hand.resources)
        Cd = calcula_BeneficiosCoste(arrayCosteCiudad, Vt)
        Cr = calcula_BeneficiosCoste(arrayCosteCarretera, Vt)
        result = GustoCiudad * Cd   - GustoPoblado * Cr *carreteraConstruible
        if result>0:
            return BuildConstants.CARD
        else:
            return BuildConstants.ROAD
        
```

```python
def borrarMaterialesSobrantes7(self, buildConstant):
    if buildConstant == BuildConstants.CITY:
        while self.hand.get_total() > 7:
            if self.hand.resources.wool > 0:
                self.hand.remove_material(4, 1)
        
            if self.hand.resources.cereal > 2:
                self.hand.remove_material(0, 1)
            if self.hand.resources.mineral > 3:
                self.hand.remove_material(1, 1)
        
            if self.hand.resources.clay > 0:
                self.hand.remove_material(2, 1)
            if self.hand.resources.wood > 0:
                self.hand.remove_material(3, 1)
    elif buildConstant == BuildConstants.TOWN:
        while self.hand.get_total() > 7:
            if self.hand.resources.wool > 1:
                self.hand.remove_material(4, 1)
        
            if self.hand.resources.cereal > 1:
                self.hand.remove_material(0, 1)
            if self.hand.resources.mineral > 0:
                self.hand.remove_material(1, 1)
        
            if self.hand.resources.clay > 1:
                self.hand.remove_material(2, 1)
            if self.hand.resources.wood > 1:
                self.hand.remove_material(3, 1)
    elif buildConstant == BuildConstants.CARD:
        while self.hand.get_total() > 7:
            if self.hand.resources.wool > 1:
                self.hand.remove_material(4, 1)
        
            if self.hand.resources.cereal > 1:
                self.hand.remove_material(0, 1)
            if self.hand.resources.mineral > 1:
                self.hand.remove_material(1, 1)
        
            if self.hand.resources.clay > 0:
                self.hand.remove_material(2, 1)
            if self.hand.resources.wood > 0:
                self.hand.remove_material(3, 1)
    elif buildConstant == BuildConstants.ROAD:
        while self.hand.get_total() > 7:
            if self.hand.resources.wool > 0:
                self.hand.remove_material(4, 1)
        
            if self.hand.resources.cereal > 0:
                self.hand.remove_material(0, 1)
            if self.hand.resources.mineral > 0:
                self.hand.remove_material(1, 1)
        
            if self.hand.resources.clay > 1:
                self.hand.remove_material(2, 1)
            if self.hand.resources.wood > 1:
                self.hand.remove_material(3, 1)
```

```python
def on_having_more_than_7_materials_when_thief_is_called(self):
    """
    Trigger que se llama cuando se debe descartar materiales. Si no los descarta el bot, los descartará
    el GameDirector aleatoriamente.
    :return: Hand()
    """
    if self.compra_objetivo is None:
       CO =  BuildConstants.CITY
    else:
        CO = self.compra_objetivo
    self.borrarMaterialesSobrantes7(CO)
    
    return self.hand
```

```python
    def on_game_start(self, board_instance):
        """
        Se llama únicamente al inicio de la partida y sirve para colocar 1 pueblo y una carretera adyacente en el mapa
        :return: int, int
        """
        #VT o VT+ o VTT+
        # Plantar en la casilla de más valor disponible 
        self.board = board_instance
        possibilities = self.board.valid_starting_nodes()
        dictNode = self.__VT__()
        maxValue=0
        bestNode=random.randint(0, 53)
        for node_id in possibilities:
            if dictNode[node_id]>maxValue:
                maxValue = dictNode[node_id]
                bestNode = node_id
        possible_roads = self.board.nodes[bestNode]['adjacent']

        return node_id, possible_roads[random.randint(0, len(possible_roads) - 1)]
```

```python
    def getValueFromGeneticVariables(self, resourcesArray):
        return resourcesArray[0]*self.cereal_gen+resourcesArray[1]*self.mineral_gen+resourcesArray[2]*self.clay_gen+ resourcesArray[3]*self.wood_gen+resourcesArray[4]*self.wool_gen
        
```
